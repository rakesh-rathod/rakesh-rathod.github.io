```python
# Two conditions on two different object type columns, use & operator
Orders[(Orders['Region']=='West') & (Orders['Sub-Category']=='Art')]
```

```python
# Two conditions on one single object type column, use | operator
Orders[(Orders['Sub-Category']=='Paper') | (Orders['Sub-Category']=='Art')]
```

```python
# Two conditions on two columns, one object type & one numeric type, use & operator
Orders[(Orders['Sub-Category']=='Tables') & (Orders['Profit'] < -500)]
```

```python
# Two conditions on two columns, both numeric type, inside/between an range/interval, use & operator
Orders[(Orders['Sales'] >= 15000) & (Orders['Sales'] <= 25000)]
```

```python
# Three conditions, two conditions on numerical columns, inside/between an range/interval, & one condition on categorical data, use & operator
Orders[(Orders['Sales'] >= 1500) & (Orders['Sales'] <= 2500) & (Orders['Sub-Category']=='Tables')]
```

```python
# Two conditions on two columns, both numeric type, outside an range/interval, use | operator
Orders[(Orders['Discount']<=0.3) | (Orders['Discount']>=0.5)]
```

```python
Shopping = pd.DataFrame()
Shopping['Product'] = ['Shirts', 'Tshirts', 'Pants', 'Belts', 'Socks', 'Shoes']
Shopping['Price/Unit'] = [1500, 800, 1500, 1000, 200, 2000]
Shopping['Quantity Purchased'] = [2,2,3,1,4,3]
Shopping['Discount'] = [0.2,0.2,0.1,0.05,0.05,0.3]
Shopping
```

```python
Shopping['MRP Sale']= (Shopping['Price/Unit'] * Shopping['Quantity Purchased'])
Shopping['Sales']= (Shopping['Price/Unit'] * Shopping['Quantity Purchased']) * (1- Shopping['Discount'])
Shopping['Discount Value 1'] = Shopping['MRP Sale'] * Shopping['Discount']
Shopping['Discount Value 2'] = Shopping['MRP Sale'] - Shopping['Sales']
Shopping
```

```python
import datetime as dt
import pandas as pd
Orders = pd.read_excel('/content/Tableau Sample SuperStore.xlsx')
Orders
OrdersDate = Orders.loc[:,['Order Date']]
OrdersDate
```

```python
OrdersDate['Order Date Year'] =OrdersDate['Order Date'].dt.strftime('%Y')
OrdersDate['Order Date Month']= OrdersDate['Order Date'].dt.strftime('%B')
OrdersDate['Order Date Month Abbr'] = OrdersDate['Order Date'].dt.strftime('%b')
OrdersDate['Order Date Weekday']= OrdersDate['Order Date'].dt.strftime('%A')
OrdersDate['Order Date Weekday Abbr'] = OrdersDate['Order Date'].dt.strftime('%a')
OrdersDate['Year2']=pd.DatetimeIndex(OrdersDate['Order Date']).year
OrdersDate['Month2']= pd.DatetimeIndex(OrdersDate['Order Date']).month
OrdersDate['Quarter']= pd.DatetimeIndex(OrdersDate['Order Date']).quarter
OrdersDate['Day'] =pd.DatetimeIndex(OrdersDate['Order Date']).day
OrdersDate
OrdersDate.dtypes
```
